```python
import pandas as pd
import nltk
from nltk import RegexpParser
# from nltk.parse.stanford import StanfordParser
# import stanfordnlp
from spacy_download import load_spacy
```

```python
nltk.download('maxent_treebank_pos_tagger')
nltk.download('treebank')
nltk.download('punkt')
nltk.download('words')
nltk.download('maxent_ne_chunker')
nltk.download('averaged_perceptron_tagger')
```

```python
df = pd.read_csv('cleaned_reviewss.csv')
```

```python
df.head(7)
```

```python
df['Review']=df['Review'].fillna("")
```

```python
df['tokenized'] = df['Review'].apply(nltk.word_tokenize)
```

```python
df['tokenized']
```

```python
df['entities'] = df['tokenized'].apply(nltk.pos_tag)
```

```python
df['entities']
```

```python
grammar_pattern = """
    NP: {<DT>?<JJ>*<NN>}
    PP: {<IN><NP>}
    VBD: {<VBD>}
    IN: {<IN>}
"""
```

```python
chunker = RegexpParser(grammar_pattern)
```

```python
df['chunks'] = df['entities'].apply(chunker.parse)
```

```python
df['chunks']
```

```python
nltk.Tree.fromstring(str(df['chunks'][770])).pretty_print()
```

```python
nlp = load_spacy('en_core_web_sm')
```

```python
deep_parse_results = []
```

```python
for sentence in df['Review']:
  doc = nlp(sentence)

  dependencies = []
  for token in doc:
    dependencies.append({
        "word":token.text,
        "lemma":token.lemma_,
        "pos":token.pos_,
        "dep":token.dep_,
        "head":token.head.text
    })

    deep_parse_results.append(dependencies)
```
